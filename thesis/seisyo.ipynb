```python
%matplotlib inline

import pandas as pd

file = pd.read_csv('WEOApr2018all.csv')
print(file.shape)
```

```python
features = []
k=0

for j in range(1,47):

        subject = file.iloc[j,1]
        feature=file.loc[file['WEO Subject Code']==subject]
        feature=feature.loc[:,'2015']

        for i in range(feature.shape[0]): # 
                feature.iloc[i]=str(feature.iloc[i])

        for i in range(feature.shape[0]): # 
                feature.iloc[i]=feature.iloc[i].replace(",","")
                feature.iloc[i]=feature.iloc[i].replace("--","nan")

        import numpy as np
        feature = feature.astype(np.float64) # 

        feature = feature.fillna(feature.median())

        features.append(feature)
        k += 1
        print(k)
        
print(features)
```

```python
print(len(features))
```

```python
keys=[]
for i in range(1,47):
        key = file.iloc[i,1]
        print(key)
        keys.append(key)
```

```python
len(keys)
```

```python
index=[]
for i in range(0,193):
        index.append(i)
print(index)
```

```python
type(features[0])
```

```python
dfeatures=pd.DataFrame({keys[0]:features[0]})
dfeatures.index=index
for i in range(1,46):
        df=pd.DataFrame({keys[i]:features[i]})
        df.index=index
        dfeatures=pd.concat([dfeatures,df],axis=1)
print(dfeatures)
```

```python
gdp=dfeatures['NGDPD']
print(gdp)
```

```python
gdp_per_capita=dfeatures['NGDPDPC']
print(gdp_per_capita)
```

```python
variables=dfeatures.drop('NGDPD',axis=1)
print(variables)
```

```python
variables_per_capita=dfeatures.drop('NGDPDPC',axis=1)
print(variables_per_capita)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# feature importance using random forest
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=80, max_features='auto')
rf.fit(variables, gdp)
print('Training done using Random Forest')

ranking = np.argsort(-rf.feature_importances_)
f, ax = plt.subplots(figsize=(11, 9))
sns.barplot(x=rf.feature_importances_[ranking], y=variables.columns.values[ranking], orient='h')
ax.set_xlabel("feature importance")
plt.tight_layout()
plt.show()
```

```python
# feature importance using random forest
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=80, max_features='auto')
rf.fit(variables_per_capita, gdp_per_capita)
print('Training done using Random Forest')

ranking = np.argsort(-rf.feature_importances_)
f, ax = plt.subplots(figsize=(11, 9))
sns.barplot(x=rf.feature_importances_[ranking], y=variables_per_capita.columns.values[ranking], orient='h')
ax.set_xlabel("feature importance")
plt.tight_layout()
plt.show()
```
